```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = .1 #some time
f=[ 1094,  1292, 780, 1161, 902, 1054, 755, 1154,1207, 1234, 760,
   1069, 951, 1086, 748, 1137,1165, 1268, 757, 1088, 905, 1147, 734, 1128, ]#f-data, enter an array
n=8 #n-number of datapoints
fhat=np.fft.fft(f,n) #f-data ,n-number of datapoints per set
freq=(1/(dt*n))*np.arange(n)
PSD2=np.log(np.abs(np.fft.fftshift(fhat))**2)
plt.plot(freq.real,PSD2.real)
plt.xlabel('frequency (Hz)')
plt.ylabel('power (dB)')
```
